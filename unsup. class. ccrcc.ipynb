```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
def kmeans_clustering_areas(area, area_indexes):
    kmeans = KMeans(n_clusters=2)
    pred = kmeans.fit_predict(area)
    indexes=area.index
    predictions=pd.DataFrame(pred, columns=['Predictions'], index=indexes)
    predictions.to_csv('K_eucl_kmeans_clustering_predictions.txt', sep='\t')
    global kmeans_tomap
    kmeans_tomap=predictions.join(area_indexes)
    return kmeans_tomap
kmeans_clustering_areas(K, K_indexes)
```

```python
#KMeans with Corr. Dist. instead of Euclidean
from sklearn.metrics import pairwise
def corrdist_kmeans_clustering_areas(area, area_indexes):
    dist=pairwise.pairwise_distances(area)
    kmeans = KMeans(n_clusters=2)
    pred = kmeans.fit_predict(dist)
    indexes=area.index
    predictions=pd.DataFrame(pred, columns=['Predictions'], index=indexes)
    predictions.to_csv('K_corrdist_kmeans_clustering_predictions.txt', sep='\t')
    global corrdist_kmeans_tomap
    corrdist_kmeans_tomap=predictions.join(area_indexes)
    return corrdist_kmeans_tomap
corrdist_kmeans_clustering_areas(K, K_indexes)
```

```python
from sklearn.cluster import AgglomerativeClustering
def aggl_clustering_areas(area, area_indexes):
    aggl = AgglomerativeClustering(n_clusters=2, linkage='ward').fit(area)
    labels = aggl.labels_
    indexes=area.index
    predictions=pd.DataFrame(labels, columns=['Predictions'], index=indexes)
    predictions.to_csv('K_aggl_clustering_predictions.txt', sep='\t')
    global aggl_tomap
    aggl_tomap=predictions.join(area_indexes)
    return aggl_tomap
aggl_clustering_areas(K, K_indexes)
```

```python
aggl_tomap=aggl_tomap.replace(0,2)
aggl_tomap=aggl_tomap.replace(1,3)
aggl_tomap=aggl_tomap.replace(2,1)
aggl_tomap=aggl_tomap.replace(3,0)
aggl_tomap
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plotting_clusters(tomap):
    fig, ax = plt.subplots()
    fig.set_size_inches(15, 9)
    plt.title('Mapping of Region K with KMeans Clustering using Eucledian Dist',fontsize=15)
    plot=sns.scatterplot(tomap['x'], tomap['y'], hue = tomap['Predictions'])
    plt.xlabel('x')
    plt.ylabel('y')
    ax.legend(fontsize=11)
    fig = plot.get_figure()
    fig.savefig("K_eucl_kmeans_clustering.pdf")
    return plt.show()
plotting_clusters(kmeans_tomap)
```
