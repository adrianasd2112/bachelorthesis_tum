```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
def kmeans_clustering_areas(area, area_indexes):
    kmeans = KMeans(n_clusters=2)
    pred = kmeans.fit_predict(area)
    indexes=area.index
    predictions=pd.DataFrame(pred, columns=['Predictions'], index=indexes)
    predictions.to_csv('O_eucl_kmeans_clustering_predictions.txt', sep='\t')
    global kmeans_tomap
    kmeans_tomap=predictions.join(area_indexes)
    return kmeans_tomap
kmeans_clustering_areas(O, O_indexes)
```

```python
#KMeans with Corr. Dist. instead of Euclidean
from sklearn.metrics import pairwise
def corrdist_kmeans_clustering_areas(area, area_indexes):
    dist=pairwise.pairwise_distances(area)
    kmeans = KMeans(n_clusters=2)
    pred = kmeans.fit_predict(dist)
    indexes=area.index
    predictions=pd.DataFrame(pred, columns=['Predictions'], index=indexes)
    predictions.to_csv('O_corrdist_kmeans_clustering_predictions.txt', sep='\t')
    global corrdist_kmeans_tomap
    corrdist_kmeans_tomap=predictions.join(area_indexes)
    return corrdist_kmeans_tomap
corrdist_kmeans_clustering_areas(O, O_indexes)
```
